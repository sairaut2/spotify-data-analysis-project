Data Cleaning

The dataset was cleaned by removing duplicate records and handling missing values using forward filling techniques. This ensures better data quality and consistency.

```python
df = df.drop_duplicates()

# Fill missing values if any
df = df.fillna(method='ffill')
```

After Data Cleaning

After cleaning, the dataset is free from major inconsistencies and is ready for further analysis and modeling.

```python
df.isnull().sum()
```

Statistical Analysis

Statistical measures such as mean, median, mode, standard deviation, variance, range, and mid-range were calculated for the popularity column.

These measures help in understanding the distribution and variability of song popularity.

```python
print("Mean:", df['popularity'].mean())
print("Median:", df['popularity'].median())
print("Mode:", df['popularity'].mode()[0])
print("Std Dev:", df['popularity'].std())
print("Variance:", df['popularity'].var())

range_val = df['popularity'].max() - df['popularity'].min()
mid_range = (df['popularity'].max() + df['popularity'].min()) / 2

print("Range:", range_val)
print("Mid-range:", mid_range)
```

Data Visualization
1. Distribution of Song Popularity

```python
plt.hist(df['popularity'], bins=20)
plt.title("Distribution of Song Popularity")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()
```

The histogram shows that most songs have moderate popularity, while only a few songs are highly popular.

2. Top Genres Distribution

```python
plt.figure(figsize=(10,5))
sns.countplot(y='genre', data=df, order=df['genre'].value_counts().index[:10])
plt.title("Top 10 Genres")
plt.show()
```

3. Popularity Distribution and Outliers

```python
sns.boxplot(x=df['popularity'])
plt.title("Popularity Spread")
plt.show()
```

4. Correlation Heatmap

```python
plt.figure()
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Simple Model

A Linear Regression model was applied to analyze the relationship between tempo and song popularity. The dataset was split into training and testing sets to evaluate model performance.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['tempo']]
y = df['popularity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("Score:", model.score(X_test, y_test))
```

The model score indicates that tempo alone does not strongly predict popularity, suggesting that multiple factors influence song success.
